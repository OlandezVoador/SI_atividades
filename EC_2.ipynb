## Desenvolvimento

```python

# Importação de biblioteca

# Acurácia do classificador
from sklearn.metrics import accuracy_score
# Matriz de confusão
from sklearn.metrics import confusion_matrix

from sklearn import cluster, decomposition
from sklearn.datasets import fetch_olivetti_faces
from sklearn.datasets import load_digits
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Função para plotar matriz de confusão
def plotar(valor):
  plt.figure(figsize=(8, 6))
  sns.heatmap(valor, annot=True, fmt="d", cmap="Blues", xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
  plt.xlabel('Predito')
  plt.ylabel('Real')
  plt.title('Matriz de Confusão')
  plt.show()

# Preparação do dataset
digits_dataset = fetch_olivetti_faces().images

# Dataset filtrado em bicondicionalidade
digits_img = digits_dataset.reshape(digits_dataset.shape[0], - 1)
digits_target = fetch_olivetti_faces().target
```

```python
#Trem de teste
from sklearn.model_selection import train_test_split as teste
x_train, x_test, y_train, y_test = teste(digits_img, digits_target, test_size=0.20)
```

```python
def rodar_modelo (metodo) :
  # TREINAMENTO DO MODELO
  metodo = metodo.fit(x_train, y_train)

  # ALGORITMO DE PREDIÇÃO DO MODELO TREINADO
  y_pred = metodo.predict(x_test)

  # Resultados
  precisao = accuracy_score(y_test, y_pred)
  print(precisao)

  precisao = confusion_matrix(y_test, y_pred)
  plotar(precisao)
```

```python
# LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis(solver='lsqr')
rodar_modelo(LDA)
```

```python
from sklearn.svm import SVC
SVM_class = SVC(C=1.0, kernel='poly', degree=3)
rodar_modelo(SVM_class)
```

```python
from sklearn.neural_network import MLPClassifier
ANN = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(50))
rodar_modelo(ANN)
```
